```
#04 - Método "winner" que busca verificar se alguém ganhou a partida no game "Quarto". 

#Função winner do game "Quarto".

#Opções de possibilidades de vitória na horizontal do tabuleiro.

#Horizontalmente.
'''
tab[0] e tab[1] e tab[2] e tab[3],
tab[4] e tab[5] e tab[6] e tab[7],
tab[8] e tab[9] e tab[10] e tab[11],  
tab[12] e tab[13] e tab[14] e tab[15]  
'''
#Método "VerificarHorizontalmente".

def VerificarHorizontalmente(list_caracteristicas,list_caracteristicas_invertidas,macro_list_pecas,tabuleiro,posicTabuleiro,contador,posictotal,qtdVezesVerificacao):
    
    contador01 = 0 
    
    contador02 = 0 
    
    #Laço de repetição principal. 

    for iprincipal in range(4):
        
       #2 laço de repetição. 

        for i in range(4):
            
            #Declarando variável que receberá a peça que está em determinada posição de uma linha horizontal do meu tabuleiro.

            verifica_pecas = tabuleiro[posicTabuleiro] - 1
            
            #Se a peça que o "verifica_pecas" está guardando for diferente de -1 que representa que naquele local não existe peça ele verifica as caracteristicas dela.

            #Vale salientar, que a 1 vez do for ele verifica a característica da cor de todas as peças na 1 linha horizontal verificada.

            if(verifica_pecas != -1):
                
                #Se existir a característica tal que a variável "list_caracteristicas" na peça que está na diagonal ele entra na condição.

                if(list_caracteristicas[iprincipal] in macro_list_pecas[verifica_pecas]):
                    
                    #Variável q conta a quantidade de peças com uma mesma característica.

                    contador01 = contador01 + 1
                    
                else:
                    
                    #Variável q conta a quantidade de peças com uma mesma característica.

                    contador02 = contador02 - 1
      
            else: #Caso o tabuleiro esteja vazio em uma determinada posição ele continua o for para outra posição da mesma linha.
                  #Observe que poderiamos otimizar o processo se colocássemos um break ao invés do continue.
                
                continue
            
            #Para a próxima vez do for ele verificar outra peça na mesma linha horizontal.

            posicTabuleiro = posicTabuleiro + 1;
        
        #A variável  "posicTabuleiro" recebe a "posictoal" que no decorrer do método guarda a posição de inicio de cada linha horizontal a ser verificada.

        posicTabuleiro = posictotal;
        
        #Se na verificação de uma característica em comum as 4 peças na horizontal tem uma mesma característica os contadores serão terão 4 ou -4.

        if(contador01 == 4):
            
            print(" QUARTO ! ! !")
            print(" FIM DE JOGO ! ! !")
            print(" Houve 4 peças posicionadas HORIZONTALMENTE com uma mesma característica, sendo a primeira característica encontrada a de serem: "+list_caracteristicas[iprincipal]+".")
            
            #Jogo acabou, portanto encerra-se o programa.
            return 1
        
        if(contador02 == -4):
            
            
            print(" QUARTO ! ! !")
            print(" FIM DE JOGO ! ! !")
            print(" Houve 4 peças posicionadas HORIZONTALMENTE com uma mesma característica, sendo a primeira característica encontrada a de serem: "+list_caracteristicas_invertidas[iprincipal]+".")
            return 1
        
        #Reinicia o contador para o a próxima verificação de outra característica da 4 peças na horizontal das linhas

        contador01 = 0
        contador02 = 0
    
    #Ao terminar os 2 for e as verificações significa q se terminou a primeira linha horizontal para se ver se as 4 peças posicionadas nela possuem características em comum.

    qtdVezesVerificacao = qtdVezesVerificacao + 1   
  
    #Tem que se verificar as 4 linhas horizontais do tabuleiro.

    if(qtdVezesVerificacao < 4):
        
        #A varíavel "posictotal" guarda a posição de inicio da próxima linha horizontal do tabuleiro.

        posictotal = 4*contador
        
        #A varíavel "posicTabuleiro" guarda o valor de "posictotal".

        posicTabuleiro = posictotal
        
        contador = contador + 1
        
        #Recursividade
        VerificarHorizontalmente(list_caracteristicas,list_caracteristicas_invertidas, macro_list_pecas,tabuleiro,posicTabuleiro,contador,posictotal,qtdVezesVerificacao)

   
#Opções de possibilidades de vitória

#Verticalmente

'''
tab[0] e tab[4] e tab[8] e tab[12], 
tab[1] e tab[5] e tab[9] e tab[13],
tab[2] e tab[6] e tab[10] e tab[14],
tab[3] e tab[7] e tab[11] e tab[15]
'''

#Método "VerificarVerticalmente", que é igualmente semelhante ao "VerificarHorizontalmente". Portanto não se faz necessário comentar cada aspecto deste método.
#Entretanto, vale salientar que neste método para se percorrer cada linha a diferença é de 4 e não de 1. 
#Além disso, a diferença no inicio de cada linha na vertical é de 1 e não de 4.

def VerificarVerticalmente(list_caracteristicas,list_caracteristicas_invertidas,macro_list_pecas,tabuleiro,posicTabuleiro,posictotal,qtdVezesVerificacao):
    
    contador01 = 0 
    
    contador02 = 0 
    
    for iprincipal in range(4):
        
        for i in range(4):
            
            verifica_pecas = tabuleiro[posicTabuleiro] - 1
            
            if(verifica_pecas != -1):
                
                if(list_caracteristicas[iprincipal] in macro_list_pecas[verifica_pecas]):
                    
                    contador01 = contador01 + 1
                   
                else:
                    
                    contador02 = contador02 - 1
                
            else:
                
                continue
            
            posicTabuleiro = posicTabuleiro + 4;
            
        posicTabuleiro = posictotal;
            
        if(contador01 == 4):
            
            print(" QUARTO ! ! !")
            print(" FIM DE JOGO ! ! !")
            print(" Houve 4 peças posicionadas VERTICALMENTE com uma mesma característica, sendo a primeira característica encontrada a de serem: "+list_caracteristicas[iprincipal]+".")
            return 1
        
        if(contador02 == -4):
            
            
            print(" QUARTO ! ! !")
            print(" FIM DE JOGO ! ! !")
            print(" Houve 4 peças posicionadas VERTICALMENTE com uma mesma característica, sendo a primeira característica encontrada a de serem: "+list_caracteristicas_invertidas[iprincipal]+".")
            return 1
        
        contador01 = 0
        contador02 = 0
    
    
    qtdVezesVerificacao = qtdVezesVerificacao + 1   
  
    
    if(qtdVezesVerificacao < 4):
        
        posictotal = posictotal + 1
        
        posicTabuleiro = posictotal
        
        VerificarVerticalmente(list_caracteristicas,list_caracteristicas_invertidas, macro_list_pecas,tabuleiro,posicTabuleiro,posictotal,qtdVezesVerificacao)

    

#Opções de possibilidades de vitória

#Diagonalmente
'''
tab[0] e tab[5] e tab[10] e tab[15], 
tab[3] e tab[6] e tab[9] e tab[12] 
'''

#Método "VerificarDiagonalmente", que é igualmente semelhante ao "VerificarHorizontalmente" e "VerificarVerticalmente". 
#Portanto não se faz necessário comentar cada aspecto deste método. 
#Entretanto, vale salientar que neste método só temos que percorrer 2 linhas diagonais. Sendo que a diferença entre posições da mesma linha de 5.
#Já na 2 linha a diferença é de 3. Por fim, observe que a diferença de inicio uma linha para outra é de 3.

def VerificarDiagonalmente(list_caracteristicas,list_caracteristicas_invertidas,macro_list_pecas,tabuleiro,posicTabuleiro,posictotal,qtdVezesVerificacao,numeroDoPulo):
    
    contador01 = 0 
    
    contador02 = 0 
    
    for iprincipal in range(4):
        
        for i in range(4):
            
            verifica_pecas = tabuleiro[posicTabuleiro] - 1
            
            if(verifica_pecas != -1):
                
                if(list_caracteristicas[iprincipal] in macro_list_pecas[verifica_pecas]):
                    
                    contador01 = contador01 + 1
                   
                else:
                    
                    contador02 = contador02 - 1
                
            else:
                
                continue
            
            posicTabuleiro = posicTabuleiro + numeroDoPulo;
            
        posicTabuleiro = posictotal;
            
        if(contador01 == 4):
            
            print(" QUARTO ! ! !")
            print(" FIM DE JOGO ! ! !")
            print(" Houve 4 peças posicionadas DIAGONALMENTE com uma mesma característica, sendo a primeira característica encontrada a de serem: "+list_caracteristicas[iprincipal]+".")
            return 1
        
        if(contador02 == -4):
            
            
            print(" QUARTO ! ! !")
            print(" FIM DE JOGO ! ! !")
            print(" Houve 4 peças posicionadas DIAGONALMENTE com uma mesma característica, sendo a primeira característica encontrada a de serem: "+list_caracteristicas_invertidas[iprincipal]+".")
            return 1
        
        contador01 = 0
        contador02 = 0
    
    
    qtdVezesVerificacao = qtdVezesVerificacao + 1   
  
    
    if(qtdVezesVerificacao < 2):
        
        numeroDoPulo = 3
        
        posictotal = 3
        
        posicTabuleiro = posictotal
        
        VerificarDiagonalmente(list_caracteristicas,list_caracteristicas_invertidas, macro_list_pecas,tabuleiro,posicTabuleiro,posictotal,qtdVezesVerificacao,numeroDoPulo)

def winner(tabuleiro):

    #Declarando as 16 peças do game "Quarto".

    list_peca01 =['branca', 'alta', 'redonda', 'com'] 
    list_peca02 =['branca', 'alta', 'quadrada', 'com'] 
    list_peca03 =['branca', 'alta', 'quadrada', 'sem'] 
    list_peca04 =['branca', 'alta', 'redonda', 'sem'] 
    list_peca05 =['preta', 'alta', 'redonda', 'sem'] 
    list_peca06 =['preta', 'alta', 'quadrada', 'com'] 
    list_peca07 =['preta', 'alta', 'quadrada', 'sem'] 
    list_peca08 =['preta', 'alta', 'redonda', 'com'] 
    list_peca09 =['preta', 'baixa', 'redonda', 'com'] 
    list_peca10 =['branca', 'baixa', 'quadrada', 'sem'] 
    list_peca11 =['branca', 'baixa', 'redonda', 'com'] 
    list_peca12 =['branca', 'baixa', 'redonda', 'sem'] 
    list_peca13 =['branca', 'baixa', 'quadrada', 'com'] 
    list_peca14 =['preta', 'baixa', 'quadrada', 'sem'] 
    list_peca15 =['preta', 'baixa', 'quadrada', 'com'] 
    list_peca16 =['preta', 'baixa', 'redonda', 'sem'] 
     
    #Criando uma lista "macro" que logo abaixo irá possuirá todas as pecas do game "Quarto".

    macro_list_pecas = [] 
     
    #Adicionando todas as pecas do game "Quarto" na lista "macro_list_pecas".

    macro_list_pecas.append (list_peca01) 
    macro_list_pecas.append (list_peca02) 
    macro_list_pecas.append (list_peca03) 
    macro_list_pecas.append (list_peca04) 
    macro_list_pecas.append (list_peca05) 
    macro_list_pecas.append (list_peca06) 
    macro_list_pecas.append (list_peca07) 
    macro_list_pecas.append (list_peca08) 
    macro_list_pecas.append (list_peca09) 
    macro_list_pecas.append (list_peca10) 
    macro_list_pecas.append (list_peca11) 
    macro_list_pecas.append (list_peca12) 
    macro_list_pecas.append (list_peca13) 
    macro_list_pecas.append (list_peca14) 
    macro_list_pecas.append (list_peca15) 
    macro_list_pecas.append (list_peca16) 

    posicTabuleiro = 0
    
    contador = 1

    posictotal = 0

    qtdVezesVerificacao = 0
    
    #Declarando uma lista que guarda características que uma peça pode possuir.

    list_caracteristicas = ['preta', 'alta', 'quadrada', 'com'] 

    #Declarando uma lista que guarda características inversas da lista "list_caracteristicas".  

    list_caracteristicas_invertidas = ['branca', 'baixa', 'redonda', 'sem'] 
    
    #Variável exclusiva para o método VerificarDiagonalmente.

    numeroDoPulo = 5 
    
    #Chamando os método "VerificarHorizontalmente", "VerificarVerticalmente","VerificarDiagonalmente" do tabuleiro passando respectivos parâmetros neles.

    #De modo que caso haja 4 peças posicionadas na horizontal, vertical e diagonal do tabuleiro com uma mesma característica o método avise e encerre o game.

    verificaVencedorHorizontalmente = VerificarHorizontalmente(list_caracteristicas,list_caracteristicas_invertidas, macro_list_pecas,tabuleiro,posicTabuleiro,contador,posictotal, qtdVezesVerificacao)

    verificaVencedorVerticalmente = VerificarVerticalmente(list_caracteristicas,list_caracteristicas_invertidas, macro_list_pecas,tabuleiro,posicTabuleiro,posictotal,qtdVezesVerificacao)
    
    verificaVencedorDiagonalmente = VerificarDiagonalmente(list_caracteristicas,list_caracteristicas_invertidas, macro_list_pecas,tabuleiro,posicTabuleiro,posictotal,qtdVezesVerificacao,numeroDoPulo)
    
    if(verificaVencedorHorizontalmente == 1 or verificaVencedorVerticalmente == 1 or verificaVencedorDiagonalmente == 1 ):
      return 1

#Utilizando método "winner".

#Tabuleiro 4x4 do jogo. Sendo que ele já está preenchido pelas peças 1, 2, 3 e 4 em sua primeira linha horizontal.

#tabuleiro = (
     
#1, 0, 0, 0, 
#0, 2, 0, 0, 
#0, 0, 3, 0, 
#0, 0, 0, 4
    
#) 

#winner(tabuleiro)


#Por fim, vale salientar que se a função quit() estivesse funcionando no colab apenas uma característica seria emitida, pois logo que houvesse as 4 peças com uma mesma
#característica o programa seria encerrado forçadamente. Assim, o tempo de execução do código consequentemente seria mais rápido.
```

```
#Foi criado as funções BotChoisePiecetoHuman e humanChoisePiecetoBot, para que o bot pudesse escolher uma peça aleatório e entregar ao jogador humano, e a outra, define a peça que vai ser escolhida ao bot,além da lista de peças inseridas, respectivamente.

#1 - Um jogador que faz jogadas ao acaso, todas as jogadas precisam ser VÁLIDAS. 
from random import randint
import random

pecasinseridas = []

def BotChoisePiecetoHuman():

  pieceToBot = randint(1,16)

  while pieceToBot in pecasinseridas:

    pieceToBot = randint(1,16)
    
  pecasinseridas.append(pieceToBot)
  
  return pieceToBot

#tab = tabuleiro
#turn = cor da peça que ele deve escolher

def humanChoisePieceToBot():

  print("Lista das peças já escolhidas:",pecasinseridas,".\n")

  pieceToHuman = int(input("Vez Bot - Escolha uma peça para o Bot: "))

  while pieceToHuman in pecasinseridas:

    print("Lista das peças já escolhidas:",pecasinseridas,".\n")

    pieceToHuman = int(input("Vez Bot - Escolha uma peça para o Bot: "))
    
  pecasinseridas.append(pieceToHuman)

  return pieceToHuman

def player(board):

  pieceBot = humanChoisePieceToBot()

  valid_moves = get_valid_moves(board)

  positionboardgoestopiece = random.choice(valid_moves)

  board[positionboardgoestopiece] = pieceBot;

  print("\nMovimentos válidos: %s.\n" % (valid_moves))

  print("Posição escolhida pelo BOT para a peça no tabuleiro: ",positionboardgoestopiece,".\n")

  print("Tabuleiro: %s.\n" % (board))

  return (random.choice(valid_moves), pieceBot)

```

#**O fluxo do jogo se dá pela seguinte sequência, onde:**
 - jogador "X" = segundo a jogar

E assim o jogo continua até que uma das condições de vitória seja obedecida.
